```python
import csv
import os
from geojson import LineString, Feature, FeatureCollection, dump
```

```python

arr = os.listdir()
print(arr)
```

```python
filelist = ['yok_okul_goc_2018.csv', 'yok_okul_goc_2016.csv', 'yok_okul_goc.csv', 'yok_okul_goc_2017.csv']
okullist = []
sehirlist = []
totallist = []
```

```python

with open('yok_okul_geo.csv', newline='') as f:
    reader = csv.reader(f)
    okullist = list(reader)
with open('yok_sehir_geo.csv', newline='') as f:
    reader = csv.reader(f)
    sehirlist = list(reader)
print(len(sehirlist), len(okullist))

for okul in okullist:
    for sehir in sehirlist:
        temp = [okul[0],okul[1],okul[2],sehir[0],sehir[1],sehir[2],0]
        totallist.append(temp)
```

```python
totallist
```

```python
count = 0
for filename in filelist:
    print(filename+' Basladi.')
    with open(filename, newline='') as f:
        reader = csv.reader(f)
        data = list(reader)
    print(data[1])
    for each in data:
        for row in totallist:
            if each[1] == row[0]:
                if each[5] == row[3]:
                    #print(each[1],row[0] , each[5] , row[3])
                    row[6]+=int(each[6])
    print(filename+' Bitti.')
print(totallist)
```

```python
# name of csv file  
filename = "yok_atlas_total.csv"
    
# writing to csv file  
with open(filename, 'w') as csvfile:  
    # creating a csv writer object  
    csvwriter = csv.writer(csvfile)  
        
    # writing the fields  
    for row in totallist:
        csvwriter.writerow(row) 
```

```python
count = 0
print(len(totallist))
geoJSON = []
for row in totallist:
    geoJSON.append(Feature(geometry=LineString([(float(row[2]), float(row[1])), (float(row[5]), float(row[4]))]),properties={"Amount": float(row[6])}))
    count+=1
    if count> 6643:
        break
feaColl = FeatureCollection(geoJSON)
#feaColl
```

```python
with open('YokAtlas_Goc.geojson', 'w') as f:
   dump(feaColl, f)
```
