```python
import csv
from geopy.geocoders import Nominatim
import time
import unicodedata
 
```

```python
with open('yok_okul_goc_2018.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)

```

```python
print(data[1][5])
```

```python
sehirlistCaps = []
for sehir in data:
    sehir[1]= sehir[1].replace(' ','')
    if not sehir[1] in sehirlistCaps:
        sehirlistCaps.append(sehir[1])
    #print(sehir[1])
sehirlistCaps
```

```python
for sehir in data:
    sehir[5]= sehir[5].replace(' ','')
    sehir[5]= sehir[5].strip()
    print(sehir[5])
```

```python
sehirlist = []
for sehir in data:
    if not sehir[5] in sehirlist:
        sehirlist.append(sehir[5])
    #print(sehir[5])
sehirlist.remove('Toplam')
sehirlist.remove('Belli Değil')
sehirlist
```

```python
sehirGeo = []
sehircapsGeo = []
control = []
for i in range(len(sehirlist)):
    control.append(False)
```

```python

while False in control:
    for i in range(len(sehirlist)):
        if control[i] == False:
            #print(each)
            geolocator = Nominatim(user_agent="ckbingol")
            location = geolocator.geocode(sehirlist[i])
            temp = [sehirlist[i] ,location.latitude, location.longitude]
            #print(location)
            print(temp)
            time.sleep(9)
            sehirGeo.append(temp)
            control[i] = True
    
#geolocator = Nominatim(user_agent="ckbingol")
#location = geolocator.geocode(sehirlist[1])
#sehirlist[1].append(location.latitude, location.longitude)
#print(location.latitude, location.longitude)
sehirGeo
```

```python

def normalize_caseless(text):
    return unicodedata.normalize("NFKD", text.casefold())

def caseless_equal(left, right):
    #print(normalize_caseless(left),normalize_caseless(right))
    return normalize_caseless(left) == normalize_caseless(right)

for sehir in sehirGeo:
    for SEHIR in sehirlistCaps:
        if caseless_equal(sehir[0],SEHIR):
            #sehir.insert(3,SEHIR)
            print('ayni')
            
sehirGeo
```
